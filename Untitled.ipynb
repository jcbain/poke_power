```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import html5lib
```

```python
def makeSoup(url):
    response=requests.get(url)
    soup=BeautifulSoup(response.content,"html.parser")
    return soup
```

```python
the_soup = makeSoup('http://pokemondb.net/type')
```

```python
weaknesses = [row['title'] for row in the_soup.findAll(['td'])]
```

```python
weaknesses[0].split('→')
```

```python
import re
re.split('→|=',weaknesses[0])
```

```python
new = []

for i in weaknesses:
    new.append(re.split('→|=',i))
```

```python
new
```

```python
import csv

with open("datasets/weaknesses.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(new)
```
